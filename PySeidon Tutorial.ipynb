### 1.4 Numpy vs. Matlab

Despite their capability similitudes, there are few major differences between *numpy* and *matlab*. Here follows probably the most important ones:
- ***[ ] vs. ( ):*** if M is a matrix, i th and j th element of M would be respectively called **M[i,j]** in *numpy* and **M(j,i)** in *matlab*.
- ***Row-major vs. Column-major:*** as one may has notice the previous point, **the order of row and column indices are inversed** between *python* and *matlab*. Consequently, be extra cautious when translating *matlab* scripts to *python* and similarly when optimisizing double for loops.
- ***0 to N-1 vs. 1 to N:*** Additionally to the index order, the number attributed to first element of a matrix is different between *python* and *matlab*. **In *matlab*, the number attributed to first element of a matrix is one wherea, in *python* it is zero**, for instance: 

```python
N = 10
M = np.ones((N)) #define a 1D array of N elements
print "Shape of M: ", M.shape
print "First element= ", M[0]
print "Last element= ", M[9]
```

- ***Indentation: *** Python uses whitespace indentation, rather than curly braces or keywords, to delimit blocks; this feature is also termed the off-side rule. An increase in indentation comes after certain statements; a decrease in indentation signifies the end of the current block...this features will be illustrated later on with exercises.

Fortunately, *numpy* and *matlab* have more similitudes than differences, so much that [*matlab* users](http://mathesaurus.sourceforge.net/matlab-numpy.html) will rapidly make their heads around *numpy*

### 2.3. Object attributs, methods & special methods

The FVCOM object possesses 4 attributs, 6 methods  (or 5 for 2D simulations) and 1 special method. They would appear by typing ***fvcomOD. Tab*** for instance.

An *attribut* is a quantity intrinsic to its *object*. A *method* is an intrinsic *function* which changes the *attributs* to its object. Contrarily a *function* will generate its own *output*:

The FVCOM attributes are:
- ***History***: history metadata that keeps track of the object changes
- ***Data***: gathers the raw/unchanged data of the specified *.nc file
- ***Grid***: gathers the grid related data
- ***Variables***: gathers the hydrodynamics related data. Note that some methods/functions will generate new fields in this attribut


The FVCOM methods & functions are:
- ***Util2D***: gathers utility methods and functions for use with 2D and 3D variables
- ***Util3D***: gathers utility methods and functions for use only with 3D variables. Note that this attribut will not appear for 2D simulations.
- ***Plots***: gathers plotting methods for use with 2D and 3D variables
- ***Harmonic_analysis***: performs a harmonic analysis on velocity components or elevation and store the so generated harmonic coefficients in ***Variables***
- ***Harmonic_reconstruction***: performs a harmonic reconstruction on velocity components or elevation and store the so reconstructed signals in ***Variables***
- ***Save_as***: save loally the current object and its attributs in a pickle file or a matlab file. Note that the so created pickle file can be use later on to define a FVCOM object and therefore restart your work where you left it...yet be careful what you wish for!!! Be aware that FVCOM runs can be very large in terms of memory when you decide to save anything from OpenDap.

## 4. PySeidon - to infinity and beyond (30 mins)

*PySeidon* can easily be coupled to any other *Python* library and package. For instance, the following script creates a series of *.png which could be then easily turn into an animated GIF with [GIMP](http://www.gimp.org/):


```python
fvcomPartial1.Util2D.hori_velo_norm()
import matplotlib.pyplot as plt
for i in range(fvcomPartial1.Grid.ntime):
    fvcomPartial1.Plots.colormap_var(fvcomPartial1.Variables.hori_velo_norm[i,:], title='Flow speed (m/s)')
    saveName = 'anim{0}.png'.format(i)
    plt.savefig(saveName, bbox_inches=0)
    plt.close()
```
