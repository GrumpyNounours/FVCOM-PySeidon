## 1. Analytical and random inputs

### 1.1. Analytical velocity field

```python
xmax = 1000.0
xmin = 0.0
Dx = 20
ymax = 500
ymin = 0.0
Dy = 20

x = np.linspace(xmin, xmax, (xmax-xmin)//Dx)
y = np.linspace(ymin, ymax,(ymax-ymin)// Dy)
X, Y = np.meshgrid(x, y)

umax = 2.5
vmax = 0.5

V = np.ones(X.shape)
for i in range(X.shape[0]):
    V[i,:] = vmax * np.cos(2.0*np.pi*(i/(X.shape[0]/2.0)))
U = umax * np.ones(Y.shape)
N = np.sqrt(np.square(U) + np.square(V))

#Plot quiver
plt.figure(figsize=(18,10))
plt.contourf(X,Y,N, alpha=.5)            
plt.colorbar()
plt.quiver(X,Y,U,V,edgecolor='k')#,facecolor='None',linewidth=.5)
```

### 1.2. Analytical control curve (identical for all device in array)

```python
cut_in=1.0
cut_out=4.5
x = np.linspace(0, 10.0, 100)
y = np.zeros(x.shape)
for i, speed in enumerate(x):
    if speed<cut_in:
        y[i] = 0.0
    else:
        y[i] = 0.2457*speed - 0.2457
    if speed>cut_out:
        y[i] = 0.86

plt.figure(figsize=(18,10))
plt.plot(x,y)
plt.title('Control Curve')
plt.ylabel("Thrust coefficient")
plt.xlabel("Inflow Speed (m/s)")
```

## 2. Wp2tidal package, step by step

### 2.1. First upload inputs through pickle files:

```python
##Velocity field
f = file(InpPath + InpName, 'rb')
data = pkl.load(f)
f.close()

##Turbines positions
f = file(InpPath + 'turb_pos.p', 'rb')
turbines = pkl.load(f)
f.close()

##Turbines features
f = file(InpPath + 'turb_fea.p', 'rb')
features = pkl.load(f)
f.close() 
```

### 2.2. Initialise "Hydro" class:

```python
hydro = Hydro(data, debug=True)
```

"hydro" gathers hydrodynamic and grid resolution components as matrices (i.e. numpy arrays). This object is representative of a "bin". The "bins" will be eventually generated by the "statistical module".

### 2.3. Initialise "Array" class:

"Array" gathers the design features (diam., position, hub velocity, control & power curves) as well as computes the streamline and relative distances of each individual device composing the array.

```python
#Plot quiver
plt.figure(figsize=(18,10))
plt.quiver(X,Y,U,V,edgecolor='k',facecolor='None',linewidth=.5)

array = Array(hydro, turbines, features, debug_plot=True)
```

### 2.4. Initialise "WakeInteraction" class:

"WakeInteraction" object gathers the induction factors and matrices (i.e. induction inter-dependencies) as the wake parametric sub-model and wake shape of each individual device composing the array. It also include the methods for resolving the induction factor matrix and wake overlapping.

```python
interaction = WakeInteraction(hydro, array)
```

### 2.5. Solve the induction factor matrix 

Through an iterative process using induction inter-dependency matrix and control curve, the "solv_induction" method resolve the array velocity interaction and update the hub velocity of each individual device composing the array.
This method shall iterate until the "convergence criteria" is met.

...bear in mind that the wake inductions are based on a "dummy" sub-model which will be eventually replace by Sandia's parametric sub-model

```python
interaction.solv_induction(convergence=0.01, debug_plot=True)
```

At this point individual device and overall array performance could be assess...yet additional aspect of the problem still need to be implemented (e.g. device yawing, blockage effects, vertical profile, turbulence effects,...)

In terms of performance, the algorithm as it is seems to run within resonable time. However, this computational time is highly sensitive to the number of devices, the device density and convergence criteria. Additionally, one must bear in mind that this computational time will be eventually multiplied by the number of "bins" and "array layout" to be tested during "hydrodynamic optimisation" as well as the computational time of the "real" parametric wake sub-model. 

```python
#Variables
bins = 6 #ebb/flood neap + ebb/flood spring + ebb/flood mean
layout = 500 #relatively small number for evolutionary optimisation algorithm for instance  
add_opt_var = 1

#Optimistic
comp_time = 10.0 #seconds
comp_time_param_sub_model = 2.0 #seconds

tot1 = (comp_time + comp_time_param_sub_model) * bins * layout * add_opt_var

#Pesimistic
comp_time = 30.0 #seconds
comp_time_param_sub_model = 10.0 #seconds

tot2 = (comp_time + comp_time_param_sub_model) * bins * layout * add_opt_var

import datetime
print "Optimistic - Total computational time: " + str(datetime.timedelta(seconds=tot1)) + " hh:mm:ss"
print "Pesimistic - Total computational time: " + str(datetime.timedelta(seconds=tot2)) + " hh:mm:ss"
```
